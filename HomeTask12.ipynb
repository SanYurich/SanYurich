2. Прочитайте данные о продажах мыла для купания за все месяцы и покажите их с помощью столбчатой диаграммы.
3. Рассчитайте общие данные о продажах за год для каждого продукта и покажите их с помощью круговой диаграммы

```python
import os
import matplotlib.pyplot as plt
root = os.getcwd()
filename = os.path.join(root, 'company_sales_data.csv')
```

```python
import pandas as pd
data = pd.read_csv(filename)
data.head()
```

```python
total_profit=data['total_profit'].tolist()
month_number=data['month_number'].tolist()
facecream=data['facecream'].tolist()
facewash=data['facewash'].tolist()
toothpaste=data['toothpaste'].tolist()
bathingsoap=data['bathingsoap'].tolist()
shampoo=data['shampoo'].tolist()
moisturizer=data['moisturizer'].tolist()
total_units=data['total_units'].tolist()
```

```python
# Прочитайте данные о продажах мыла для купания за все месяцы и покажите их с помощью столбчатой диаграммы.
plt.figure(figsize =(10, 10))
plt.grid(linewidth=4, linestyle=':')
plt.bar([i-1 for i in month_number], facecream, label ='bathingsoap', align='center',width=0.8)
plt.xticks(month_number)
plt.title('Year_bathingsoap')
plt.xlabel('month_number')
plt.ylabel('units')
plt.legend(loc = 'upper right')
plt.show()
```

```python
# Рассчитайте общие данные о продажах за год для каждого продукта и покажите их с помощью круговой диаграммы

plt.figure(figsize =(10, 10)) 
plt.plot(month_number, moisturizer,'r--',label ='moisturizer')
plt.plot(month_number, shampoo, 'g-',label ='shampoo')
plt.plot(month_number, bathingsoap, 'b:',label ='bathingsoap')
plt.plot(month_number, toothpaste, 'c-.',label ='toothpaste')
plt.plot(month_number, facewash, 'm.',label ='facewash')
plt.plot(month_number, facecream, 'y',label ='facecream')
plt.title('Sale of products')
plt.xlabel('Months')
plt.ylabel('Unit')
plt.xticks(month_number)
plt.legend(loc = 'upper left')
plt.show()
```

```python
#Рассчитайте общие данные о продажах за год для каждого продукта и покажите их с помощью круговой диаграммы

facecream_total=sum(facecream)
facewash_total=sum(facewash)
toothpaste_total=sum(toothpaste)
bathingsoap_total=sum(bathingsoap)
shampoo_total=sum(shampoo)
moisturizer_total=sum(moisturizer)

plt.figure(figsize =(10, 10)) 
labels = ['facecream_total','facewash_total','toothpaste_totale','bathingsoap_total','shampoo_total', 'moisturizer_total']
values = [facecream_total,facewash_total,toothpaste_total,bathingsoap_total, shampoo_total, moisturizer_total]
colors = ['yellow','green','red','blue','black','pink']
plt.pie(values,labels=labels,colors=colors)
plt.axis('equal')
plt.show()
```
